```python
def create_profile(name, topics):
  text = wiki_wiki.page(name).sections
  profile = dict()

  for topic in topics:
    profile[topic] = []

  subsections = str(text).split('Section')

  for topic in topics:
    for j in range(len(subsections)):
      if topic in str(subsections[j]):
        profile[topic].append(subsections[j])
  return profile
```

```python
def grade_profile(profile, topics):
  grade = 0
  topics_evaluated = 0
  full_report = dict()
  for topic in topics:
    topic_results = []
    texts = profile[topic]
    if texts != []:
      for i, text in enumerate(texts):
          model_name = models[i]
          inference_result = inference(text, model_name, tokenizer)
          if inference_result[0] > 0.90:
            topic_results.append(0)
          elif inference_result[1] > 0.6:
            topic_results.append(1)
          else:
            pass

      if topic_results.count(0) > topic_results.count(1):
        verdict  = 0
        topics_evaluated += 1
        grade += verdict
        full_report[topic] = verdict

      elif topic_results.count(0) < topic_results.count(1):
        verdict = 1
        topics_evaluated += 1
        grade += verdict
        full_report[topic] = verdict

      else:
        pass
  try:
    return (grade/topics_evaluated)*10, full_report
  except:
    return "NO INFO WAS FOUND", "NO INFO WAS FOUND"
```

```python
def inference(text, model, tokenizer):
  inputs = tokenizer(text, truncation=True, padding=True, return_tensors="pt").to(device)
  with torch.no_grad():
    logits = model(**inputs).logits
  
  logi = []
  for i in range(len(logits)):
      logi.append(logits[i].cpu())
  t1, t2 = softmax(logi[0])[0], softmax(logi[0])[1]
  return [t1, t2]
```

```python
def batch_grading(names):
  result = []
  reports = []
  for name in names:
    profile = create_profile(name, topics)
    grade, full_report = grade_profile(profile, topics)
    result.append(grade)
    reports.append(full_report)
  return result, reports
```

```python
def test_models(names, label, topics):
  models_results = dict()
  return_pr = dict()
  for topic in topics:
    models_results[topic] = [0, 0]
  for name in names:
    profile = create_profile(name, topics)
    grade, full_report = grade_profile(profile, topics)
    try:
      for key in list(full_report.keys()):
        if full_report[str(key)] == label:
          models_results[str(key)][0] += 1
        else:
          models_results[str(key)][1] += 1
    except:
      pass

  for key in list(models_results.keys()):
    try:
      return_pr[key] = models_results[str(key)][0] / (models_results[str(key)][0] + models_results[str(key)][1] )
    except:
      pass

  return return_pr , models_results
```
